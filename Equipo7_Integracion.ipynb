```python
#Nos quedamos únicamente con el conjunto de columnas del coneval que queremos: 'clave_entidad', 'municipio','ic_sbv_pob', 'pobreza_pob'
coneval_raw = coneval_raw[['clave_entidad', 'municipio','ic_sbv_pob', 'pobreza_pob']]
```

```python
#Nos quedamos únicamente con el total de cada municipio (ignoramos localidades)
inegi_raw = inegi_raw[inegi_raw["NOM_LOC"].str.strip() == "Total del Municipio"].copy()
#Para hacer el merge necesitamos renombrar algunas columnas
inegi_raw.rename(columns={'NOM_MUN': 'municipio', 'ï»¿ENTIDAD': 'clave_entidad'}, inplace=True)
```

```python
inegi_raw.head()
```

```python
#hacemos el merge
df_raw = pd.merge(inegi_raw,coneval_raw, on=['municipio','clave_entidad'], how='left')
```
